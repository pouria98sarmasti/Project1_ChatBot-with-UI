```python
from datasets import load_dataset
from transformers import AutoTokenizer

from langchain_core.documents import Document
from langchain_text_splitters.character import CharacterTextSplitter, RecursiveCharacterTextSplitter
from langchain_community.vectorstores.faiss import FAISS
from langchain_huggingface.embeddings.huggingface import HuggingFaceEmbeddings

embedding_e5 = HuggingFaceEmbeddings(
                            model_name="heydariAI/persian-embeddings",
                            model_kwargs={'device': 'cuda'},
                            encode_kwargs={'normalize_embeddings': False}
                            )

drugQA = load_dataset("amirmmahdavikia/darooyab_qa", split="train")
```

```python
def remove_rows_with_no_question(example):
  return example['question'] is not None

drugQA_removedRows_withNoQuestion = drugQA.filter(remove_rows_with_no_question)
```

```python
from hazm import InformalNormalizer, Normalizer

normalizer = Normalizer()
informal_normalizer = InformalNormalizer()



def normalize_text(example):

  normalized_question = normalizer.normalize(example["question"])
  normalized_answer = normalizer.normalize(example["answer"])

  return {"question": normalized_question,
          "answer": normalized_answer}


# drugQA_normalized = drugQA_removedRows_withNoQuestion.map(normalize_text, batched=True)  #### batched=True not working
drugQA_normalized = drugQA_removedRows_withNoQuestion.map(normalize_text)
```

```python
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=200, separator=[" "], is_separator_regex=False,)
```

```python
document_drugQA = [ Document(page_content=qa["question"], metadata={"answer": qa["answer"]})\
                   for qa in drugQA_normalized]
# document_drugQA = [f"question: {qa['question']} \n answer: {qa['answer']}" for qa in drugQA_normalized]
# doucument_drugQA = text_splitter.split_documents(document_drugQA)

vectorDB_faiss_drugQA = FAISS.from_documents(document_drugQA, embedding_e5)
```

###workflow: for the models that support bind_tools

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import BaseMessage
from langchain_core.tools.convert import tool
from langchain_core.runnables.base import chain

from langgraph.graph.message import add_messages
from langgraph.graph import START, END, StateGraph
from langgraph.checkpoint.memory import MemorySaver
from langgraph.prebuilt import ToolNode, tools_condition

from typing import Sequence
from typing_extensions import Annotated, TypedDict

from pydantic import BaseModel, Field

from sqlalchemy import create_engine, text
```

```python
class OutputStructure_CriticismIdentifier(BaseModel):
    is_critisism: str = Field(description="is statement critical?", enum=["yes", "no"])
    # is_critisism: bool = Field(description="is statement critical?", enum=["True", "False"])
    # aggressiveness: int = Field(description="describes how aggressive the statement is, the higher the number the more aggressive", enum=["1", "2", "3", "4", "5"])
    # language: str = Field(description="the language of statement", enum=["persian", "english"])


class State(TypedDict):
  messages: Annotated[Sequence[BaseMessage], add_messages] # include all messages
  conversation: Annotated[Sequence[BaseMessage], add_messages] # only system, human, final ai messages
  is_user_input_criticism: OutputStructure_CriticismIdentifier
  language: str

```

```python
# @tool(response_format="content_and_artifact")
@tool
def retriever(query: str):
  """Retrieve information related to questions about medicine."""

  retrieved_docs = vectorDB_faiss_drugQA.similarity_search(query, k=1)

  return retrieved_docs


def add_conversation_to_messages(state: State):
  return {"messages": [state["conversation"][-1]]}


def criticism_identifier(state: State):

  llm_criticism_identifier = llm.with_structured_output(OutputStructure_CriticismIdentifier)

  promptSystem_CriticismIdentifer = """Extract the desired information from the input statement.\n
  Only extract the properties mentioned in the 'OutputStructure_CriticismIdentifier' function."""

  chatPromptTemplate_CriticismIdentifer = ChatPromptTemplate([("system", promptSystem_CriticismIdentifer),
                                                            ("human", "{input}"),
                                                            ])
  chain_criticism_identifier = chatPromptTemplate_CriticismIdentifer | llm_criticism_identifier

  return {"is_user_input_criticism": chain_criticism_identifier.invoke(state["conversation"][-1].content)}

# conditional edge
def does_store_in_sql(state: State):
  is_criticism = state["is_user_input_criticism"].dict()["is_critisism"]
  if is_criticism:
    return "write_criticism_to_db"
  else:
    return "respond_or_retrieve"


###### chain
def write_criticism_to_db(state: State):

  is_user_input_criticism = state["is_user_input_criticism"].dict()
  engine = create_engine("sqlite:///criticismDB.db")

  with engine.connect() as conn:
    try:
      # conn.execute(text("CREATE TABLE table_criticism (criticism string, aggressiveness int, language string)"))
      conn.execute(text("CREATE TABLE table_criticism (criticism text)"))
    finally:
      # conn.execute(text("INSERT INTO table_criticism (criticism) VALUES (:criticism, :aggressiveness, :language)"), [{"criticism_comment": state["conversation"][-1].content,
      #                                                                                                               "aggressiveness": is_user_input_criticism["aggressiveness"],
      #                                                                                                               "language": is_user_input_criticism["language"],
      #                                                                                                               }])
      conn.execute(text("INSERT INTO table_criticism (criticism) VALUES (:criticism, :aggressiveness, :language)"), [{"criticism_comment": state["conversation"][-1].content}])
      conn.commit()


def respond_or_retrieve(state: State):

  llm_with_tools = llm.bind_tools([retriever])
  response = llm_with_tools.invoke(state["messages"])
  return {"messages": [response]}



def generate(state: State):

  recent_tool_messages = []
  for message in reversed(state["messages"]):
      if message.type == "tool":
          recent_tool_messages.append(message)
      else:
          break
  tool_messages = recent_tool_messages[::-1]
  tool_messages_content = "\n\n".join(tool_message.content for tool_message in tool_messages)

  promptTemplate_system = """
  1. You are an assistant for question-answering tasks related to medicines.
  2. Use the following pieces of retrieved context to answer the question.
  3. If you don't know the answer, say that you don't know.
  4. Use three sentences maximum and keep the answer concise.
  5. Always answer in {language}.
  "\n\n"
  retrieved context: {context}
  """
  promptTemplate_system = PromptTemplate.from_template(promptTemplate_system)

  prompt_system = promptTemplate_system.format(language=state["language"], context=tool_messages_content)

  chatPromptTemplate = ChatPromptTemplate([
      ("system", prompt_system),
      ("placeholder", "{conversation}"), # or use: MessagesPlaceholder(variable_name="conversation", optional=True)
  ])

  prompt_chat = chatPromptTemplate.format(conversation=state["conversation"])


  response = llm.invoke(prompt_chat)

  return {"conversation": [response], "messages": [response]}

retriever_node = ToolNode([retriever], messages_key="messages")
```

```python
app_chatBot_builder = StateGraph(state_schema=State)

app_chatBot_builder.add_node("add_conversation_to_messages", add_conversation_to_messages)
app_chatBot_builder.add_node("criticism_identifier", criticism_identifier)
app_chatBot_builder.add_node("write_criticism_to_db", write_criticism_to_db)
app_chatBot_builder.add_node("respond_or_retrieve", respond_or_retrieve)
app_chatBot_builder.add_node("retriever_node", retriever_node)
app_chatBot_builder.add_node("generate", generate)

# app_chatBot_builder.set_entry_point("add_conversation_to_messages")
app_chatBot_builder.add_edge(START, "add_conversation_to_messages")
app_chatBot_builder.add_edge("add_conversation_to_messages", "criticism_identifier")
app_chatBot_builder.add_conditional_edges("criticism_identifier", does_store_in_sql)
app_chatBot_builder.add_edge("write_criticism_to_db", "respond_or_retrieve")
app_chatBot_builder.add_conditional_edges("respond_or_retrieve", tools_condition, {END: END, "tools": "retriever_node"})
app_chatBot_builder.add_edge("retriever_node", "generate")
app_chatBot_builder.add_edge("generate", END)


memory = MemorySaver()
app_chatBot = app_chatBot_builder.compile(checkpointer=memory)
```

###workflow: for the models that does not support bind_tools

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import BaseMessage
from langchain_core.tools.convert import tool
from langchain_core.runnables.base import chain

from langgraph.graph.message import add_messages
from langgraph.graph import START, END, StateGraph
from langgraph.checkpoint.memory import MemorySaver
from langgraph.prebuilt import ToolNode, tools_condition

from typing import Sequence, Any, Dict
from typing_extensions import Annotated, TypedDict

from pydantic import BaseModel, Field

from sqlalchemy import create_engine, text
```

```python
class OutputStructure_CriticismIdentifier(BaseModel):
    is_critisism: str = Field(description="is statement critical?", enum=["yes", "no"])
    # is_critisism: bool = Field(description="is statement critical?", enum=["True", "False"])
    # aggressiveness: int = Field(description="describes how aggressive the statement is, the higher the number the more aggressive", enum=["1", "2", "3", "4", "5"])
    # language: str = Field(description="the language of statement", enum=["persian", "english"])


class State(TypedDict):
  conversation: Annotated[Sequence[BaseMessage], add_messages] # only system, human, final ai messages
  is_user_input_criticism: OutputStructure_CriticismIdentifier
  language: str
  call_retriever: str

```

```python
# @tool(response_format="content_and_artifact")
### @tool
def retriever(query: str):
  """Retrieve information related to questions about medicine."""

  retrieved_docs = vectorDB_faiss_drugQA.similarity_search(query, k=1)

  concatenated_docs = "\n\n".join(f"question: {doc.page_content} answer: {doc.metadata['answer']}" for doc in  retrieved_docs)

  return concatenated_docs



def criticism_identifier(state: State):

  promptSystem_CriticismIdentifer = """Extract the desired information from the input statement.\n
  Only extract the properties mentioned in the 'OutputStructure_CriticismIdentifier' function."""

  chatPromptTemplate_CriticismIdentifer = ChatPromptTemplate([("system", promptSystem_CriticismIdentifer),
                                                            ("human", "{input}"),
                                                            ])
  chain_criticism_identifier = chatPromptTemplate_CriticismIdentifer | llm.with_structured_output(OutputStructure_CriticismIdentifier)

  return {"is_user_input_criticism": chain_criticism_identifier.invoke(state["conversation"][-1].content)}



# conditional edge
def does_store_in_sql(state: State):
  is_criticism = state["is_user_input_criticism"].dict()["is_critisism"]
  if is_criticism=="yes":
    return "write_criticism_to_db"
  else:
    return "respond_or_retrieve"


###### chain
def write_criticism_to_db(state: State):

  is_user_input_criticism = state["is_user_input_criticism"].model_dump()
  engine = create_engine("sqlite:///criticismDB.db")

  with engine.connect() as conn:
    try:
      # conn.execute(text("CREATE TABLE table_criticism (criticism string, aggressiveness int, language string)"))
      conn.execute(text("CREATE TABLE table_criticism (criticism text)"))
    finally:
      # conn.execute(text("INSERT INTO table_criticism (criticism) VALUES (:criticism, :aggressiveness, :language)"), [{"criticism_comment": state["conversation"][-1].content,
      #                                                                                                               "aggressiveness": is_user_input_criticism["aggressiveness"],
      #                                                                                                               "language": is_user_input_criticism["language"],
      #                                                                                                               }])
      conn.execute(text("INSERT INTO table_criticism (criticism) VALUES (:criticism, :aggressiveness, :language)"), [{"criticism_comment": state["conversation"][-1].content}])
      conn.commit()


def respond_or_retrieve(state: State):

  class IsRelatedToMedicine(BaseModel):
    relation: str = Field(..., description="is input related to medicine field.", enum=["yes", "no"])

  llm_with_tool = llm.with_structured_output(IsRelatedToMedicine)
  llm_out = llm_with_tool.invoke(state["conversation"][-1])

  return {"call_retriever": llm_out.relation}




# conditional edge
def does_call_retriever(state: State):

  if state["call_retriever"]=="yes":
    return "generate_with_tool"
  else:
    return "generate_without_tool"




def generate_with_tool(state: State):

  tool_messages_content = retriever(state["conversation"][-1].content)

  promptTemplate_system = """
  1. You are an assistant for question-answering tasks related to medicines.
  2. Use the following pieces of retrieved context to answer the question.
  3. If you don't know the answer, say that you don't know.
  4. Use three sentences maximum and keep the answer concise.
  5. Always answer in {language}.
  "\n\n"
  retrieved context: {context}
  """
  promptTemplate_system = PromptTemplate.from_template(promptTemplate_system)

  prompt_system = promptTemplate_system.format(language=state["language"], context=tool_messages_content)

  chatPromptTemplate = ChatPromptTemplate([
      ("system", prompt_system),
      ("placeholder", "{conversation}"), # or use: MessagesPlaceholder(variable_name="conversation", optional=True)
  ])

  prompt_chat = chatPromptTemplate.format(conversation=state["conversation"])

  response = llm.invoke(prompt_chat)

  return {"conversation": [response]}


def generate_without_tool(state: State):

  promptTemplate_system = """
  1. You are an assistant for question-answering tasks related to medicines.
  2. Always answer in {language}.
  """
  promptTemplate_system = PromptTemplate.from_template(promptTemplate_system)

  prompt_system = promptTemplate_system.format(language=state["language"])

  chatPromptTemplate = ChatPromptTemplate([
      ("system", prompt_system),
      ("placeholder", "{conversation}"), # or use: MessagesPlaceholder(variable_name="conversation", optional=True)
  ])

  prompt_chat = chatPromptTemplate.format(conversation=state["conversation"])

  response = llm.invoke(prompt_chat)

  return {"conversation": [response]}
```

```python
app_chatBot_builder = StateGraph(state_schema=State)

app_chatBot_builder.add_node("criticism_identifier", criticism_identifier)
app_chatBot_builder.add_node("write_criticism_to_db", write_criticism_to_db)
app_chatBot_builder.add_node("respond_or_retrieve", respond_or_retrieve)
app_chatBot_builder.add_node("generate_with_tool", generate_with_tool)
app_chatBot_builder.add_node("generate_without_tool", generate_without_tool)

app_chatBot_builder.set_entry_point("criticism_identifier")
app_chatBot_builder.add_conditional_edges("criticism_identifier", does_store_in_sql)
app_chatBot_builder.add_edge("write_criticism_to_db", "respond_or_retrieve")
app_chatBot_builder.add_conditional_edges("respond_or_retrieve", does_call_retriever)
app_chatBot_builder.add_edge("generate_with_tool", END)
app_chatBot_builder.add_edge("generate_without_tool", END)


memory = MemorySaver()
app_chatBot = app_chatBot_builder.compile(checkpointer=memory)
```
