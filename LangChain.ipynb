```python
# step1: create a chat model
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(**openai_kwargs)
```

```python
from langchain_core.messages import HumanMessage
from langgraph.checkpoint.memory import MemorySaver
from langgraph.graph import START, MessagesState, StateGraph
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from typing import Sequence
from langchain_core.messages import BaseMessage
from langgraph.graph.message import add_messages
from typing_extensions import Annotated, TypedDict


# step 2: create ChatPtomptTemplate and runnable
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "Answer in {language}."),
        MessagesPlaceholder(variable_name="messages"), # variable_name is list of messages
    ]
)
runnable = prompt | llm

# step 3: initialize StateGraph with its proper state_schema
class State(TypedDict):
    messages: Annotated[Sequence[BaseMessage], add_messages]
    language: str

workflow = StateGraph(state_schema=State)


# step 4: add_edge and add_node to StateGraph
workflow.add_edge(START, "model")

def call_model(state: State):
    response = runnable.invoke(state)
    # Update message history with response:
    return {"messages": [response]}

workflow.add_node("model", call_model)


# step 5: compile StateGraph with memory
memory = MemorySaver()
app = workflow.compile(checkpointer=memory)
```

```python
%%writefile cv_summarizer.py

openai_api_key = 'ollama'
BASE_URL = 'http://localhost:11434/v1/'

openai_kwargs = {'openai_api_key': openai_api_key,
                 'base_url': BASE_URL,
                 'max_tokens': 100,
                 'model': 'llama3.2',
                 'temperature': 1}


from langchain_community.document_loaders.pdf import PyPDFLoader
from langchain_community.document_loaders.word_document import Docx2txtLoader
from langchain_text_splitters.character import CharacterTextSplitter
from langchain_core.prompts.prompt import PromptTemplate
from langchain.chains.summarize.chain import load_summarize_chain
from langchain_openai import ChatOpenAI
import streamlit as st
from pathlib import Path


def process_docx(docx_file):
    text=""
    loader=Docx2txtLoader(docx_file)
    texts = loader.load_and_split()
    return texts


def process_pdf(pdf_file):
    text=""
    #PYPDFLoader loads a list of PDF Document objects
    loader=PyPDFLoader(pdf_file)
    pages = loader.load()

    for page in pages:
        text+=page.page_content
    text= text.replace('\t', ' ')


    text_splitter = CharacterTextSplitter(
            separator="\n",
            chunk_size=1000,
            chunk_overlap=50
        )
    texts = text_splitter.create_documents([text])

    return texts


def main():

    st.title("CV Summary Generator")

    uploaded_file = st.file_uploader("Select CV", type=["docx", "pdf"])
    # uploaded_file = Path("/content/Software-Engineer-CV.pdf")

    text = ""
    if uploaded_file is not None:
        file_extension = uploaded_file.name.split('.')[-1]

        st.write("File Details:")
        st.write(f"File Name: {uploaded_file.name}")
        st.write(f"File Type: {file_extension}")

        # step 1: load the document
        if file_extension == "docx":
            text = process_docx(uploaded_file.name)
        elif file_extension == "pdf":
            text = process_pdf(uploaded_file.name)
        else:
            st.error("Unsupported file format. Please upload a .docx or .pdf file.")
            return
        llm = ChatOpenAI(**openai_kwargs)
        prompt_template = """You have been given a Resume to analyse.
        Write a verbose detail of the following:
        {text}
        Details:"""
        prompt = PromptTemplate.from_template(prompt_template)

        refine_template = (
            "Your job is to produce a final outcome\n"
            "We have provided an existing detail: {existing_answer}\n"
            "We want a refined version of the existing detail based on initial details below\n"
            "------------\n"
            "{text}\n"
            "------------\n"
            "Given the new context, refine the original summary in the following manner:"
            "Name: \n"
            "Email: \n"
            "Key Skills: \n"
            "Last Company: \n"
            "Experience Summary: \n"
        )
        refine_prompt = PromptTemplate.from_template(refine_template)

        # step 2: load_summarize_chain with chain_type=refine
        chain_cv = load_summarize_chain(
            llm=llm,
            chain_type="refine",
            question_prompt=prompt,
            refine_prompt=refine_prompt,
            return_intermediate_steps=True,
            input_key="input_documents",
            output_key="output_text",
        )
        result = chain_cv({"input_documents": text}, return_only_outputs=True)

        st.write("Resume Summary:")
        st.text_area("Text", result['output_text'], height=400)


# if __name__ == "__main__":
#     main()

main()
```

###Project 3: invoice data extractor

```python
%%writefile invoice_data_extractor.py

openai_api_key = 'ollama'
BASE_URL = 'http://localhost:11434/v1/'

openai_kwargs = {'openai_api_key': openai_api_key,
                 'base_url': BASE_URL,
                 'max_tokens': 1000,
                 'model': 'llama3.2',
                 'temperature': 0}

from langchain_community.document_loaders.pdf import PyPDFLoader
from langchain_huggingface.embeddings.huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores.faiss import FAISS
from langchain_core.prompts.prompt import PromptTemplate
from langchain.chains.combine_documents.stuff import create_stuff_documents_chain
from langchain.chains.retrieval import create_retrieval_chain
from langchain_openai import ChatOpenAI
import streamlit as st
import json
import pandas as pd

embeddings_hf = HuggingFaceEmbeddings(
                            model_name="intfloat/multilingual-e5-large",
                            model_kwargs={'device': 'cuda'},
                            encode_kwargs={'normalize_embeddings': False}
                            )

def create_docs(user_pdf_list):
    """This function is used to extract the invoice data from the given PDF files."""
    answer = []
    for filename in user_pdf_list:

        loader = PyPDFLoader(filename.name)
        pages = loader.load_and_split()
        vectorDB_faiss = FAISS.from_documents(pages, embeddings_hf)

        template = """Extract all the following values : invoice no., Description, Quantity, date,
            Unit price , Amount, Total, email, phone number and address
            from the following Invoice content
            (create a JSON output with the extracted fields only):
            {context}
            The fields and values in the above content may be jumbled up as they are extracted from a PDF. Please use your judgement to align
            the fields and values correctly based on the fields asked for in the question above.
            Expected output JSON format as follows:
            {{"Invoice no.": "xxxxxxxx","Description": "xxxxxx","Quantity": "x","Date": "dd/mm/yyyy",
            "Unit price": "xxx.xx","Amount": "xxx.xx","Total": "xxx,xx","Email": "xxx@xxx.xxx","Phone number": "xxxxxxxxxx","Address": "xxxxxxxxx"}}
            Remove any dollar symbols or currency symbols from the extracted values.
            Your output must only contain JSON format section wihtout any extra explanation.
            ensure that your output is proper to be used with "json.loads" function.
            fields of json file that not any provided data, or ambigious data, must left empty. like this: "empty field name": ""
            """
        prompt_documents = PromptTemplate.from_template(template)

        llm = ChatOpenAI(**openai_kwargs)
        retriever_faiss = vectorDB_faiss.as_retriever()
        chain_document = create_stuff_documents_chain(llm, prompt_documents)
        chain_retrieval = create_retrieval_chain(retriever_faiss, chain_document)
        response = chain_retrieval.invoke({"input": ""})
        answer.append(response['answer'])

    return pd.DataFrame([json.loads(row) for row in answer])


def main():

    st.set_page_config(page_title="Invoice Extraction Bot")
    st.title("Invoice Extraction Bot...💁 ")
    st.subheader("I can help you in extracting invoice data")

    pdf = st.file_uploader("Upload invoices here, only PDF files allowed", type=["pdf"],accept_multiple_files=True)
    submit=st.button("Extract Data")

    if submit:
        with st.spinner('Wait for it...'):
            ExtractedData_DFFormat = create_docs(pdf)
            st.write(ExtractedData_DFFormat)

        st.success("Hope I was able to save your time❤️")


if __name__ == '__main__':
    main()
```

###Project 5: summarize very long text using MapReduce Technique

```python
from langchain_core.output_parsers.string import StrOutputParser
from langchain_core.prompts.chat import ChatPromptTemplate

from langchain_community.document_loaders.web_base import WebBaseLoader
from langchain_text_splitters.character import CharacterTextSplitter

import operator
from typing import Annotated, List, Literal, TypedDict

from langchain.chains.combine_documents.reduce import acollapse_docs, split_list_of_docs
from langchain_core.documents.base import Document

from langgraph.constants import Send
from langgraph.graph import END, START, StateGraph
```

```python
llm = ChatOpenAI(**openai_kwargs)
```

```python
loader = WebBaseLoader("https://docs.llamaindex.ai/en/stable/examples/pipeline/query_pipeline_sql/")
docs = loader.load()

text_splitter = CharacterTextSplitter.from_tiktoken_encoder(chunk_size=1000, chunk_overlap=0)
split_docs = text_splitter.split_documents(docs)

print(f"Generated {len(split_docs)} documents.")
```

```python
prompt_map_summary = ChatPromptTemplate([("system", "Write a concise summary of the input."),
                                         ("human", "{context}")])

prompt_system_final_summary = """
You will be provided a set of summaries.
Take these and distill it into a final, consolidated summary
of the main themes.
"""
prompt_final_summary = ChatPromptTemplate.from_messages([("system", prompt_system_final_summary),
                                                         ("human", "The following is a set of summaries: {docs}")])

chain_map_summary = prompt_map_summary | llm | StrOutputParser()
chain_final_summary = prompt_final_summary | llm | StrOutputParser()
```

```python
TOKEN_MAX = 1000

class OveralState(TypedDict):
  contents: List[str]
  summaries: Annotated[list, operator.add] # operator.add is essintial for the return of the node (function) that is mapped
  collapsed_summaries: List[Document]
  final_summary: str

# the state for the input of the mapped node
class SummaryState(TypedDict):
  content: str


def length_function(documents: List[Document]) -> int:
  """Get number of tokens for input contents."""
  return sum(llm.get_num_tokens(doc.page_content) for doc in documents)


async def generate_summary(state: SummaryState) -> OveralState:
  response = await chain_map_summary.ainvoke(state["content"])
  return {"summaries": [responce]}


# the edge for MapReduce technique (using Send class)
def map_summary(state: OveralState):
  return [Send("generate_summary", {"content": content}) for content in state["contents"]]

# we need a List[Document] instead of List[str]. so we can apply collapse function
def collect_summaries(state: OveralState):
  return {"collapsed_summaries": [Document(summary) for summary in state["summaries"]]}


async def collapse_summaries(state: OverallState):
    doc_lists = split_list_of_docs(state["collapsed_summaries"], length_function, token_max)
    results = []
    for doc_list in doc_lists:
        results.append(await acollapse_docs(doc_list, chain_final_summary.ainvoke))
    return {"collapsed_summaries": results}


def should_collapse(state: OveralState) -> Literal["collapse_summaries", "generate_final_summary"]:
  num_tokens = length_function(state["collapsed_summaries"])
  if num_tokens > token_max:
      return "collapse_summaries"
  else:
      return "generate_final_summary"



async def generate_final_summary(state: OverallState):
    response = await reduce_chain.ainvoke(state["collapsed_summaries"])
    return {"final_summary": response}


# Construct the graph
# Nodes:
graph_summarizer = StateGraph(state_schema=OverallState)
graph_summarizer.add_node("generate_summary", generate_summary)  # same as before
graph_summarizer.add_node("collect_summaries", collect_summaries)
graph_summarizer.add_node("collapse_summaries", collapse_summaries)
graph_summarizer.add_node("generate_final_summary", generate_final_summary)

# Edges:
graph_summarizer.add_conditional_edges(START, map_summaries, ["generate_summary"])
graph_summarizer.add_edge("generate_summary", "collect_summaries")
graph_summarizer.add_conditional_edges("collect_summaries", should_collapse)
graph_summarizer.add_conditional_edges("collapse_summaries", should_collapse)
graph_summarizer.add_edge("generate_final_summary", END)

app_summarizer = graph_summarizer.compile()
```

```python
from typing import Any, Dict, Optional, TypedDict
from pydantic import BaseModel, Field
```

```python
from langchain_core.tools.convert import tool
from langchain_core.tools.render import render_text_description, render_text_description_and_args
from langchain_core.prompts.chat import ChatPromptTemplate
from langchain_core.runnables.passthrough import RunnablePassthrough
from langchain_openai.chat_models.base import ChatOpenAI
from langchain_core.runnables.config import RunnableConfig
```

```python
llm = ChatOpenAI(**openai_kwargs)
```

```python
@tool
def add(a: float, b: float) -> float:
    """Adds a and b."""
    return a + b


@tool
def multiply(a: float, b: float) -> float:
    """Multiplies a and b."""
    return a * b


tools = [add, multiply]
tool_names = [tool.name for tool in tools]
```

```python
class OutputedTool(BaseModel):
  name: str = Field(..., description="name of the tool that must be used", enum=tool_names) # Field(..., ) means that we do not set default value (vs Field('name', ..., ))
  args: Dict[str, Any] = Field(..., description="arguments of the tool and their values.")


def invoke_tool(tool_call_request: OutputedTool, config: Optional[RunnableConfig] = None):

    tool_name_to_tool = {tool.name: tool for tool in tools}
    name = tool_call_request.name
    requested_tool = tool_name_to_tool[name]
    return requested_tool.invoke(tool_call_request.args, config=config)
```

```python
llm_with_tool = llm.with_structured_output(OutputedTool) | invoke_tool
```
