#chatbot

###ollama

```python
import ollama
response: ollama.ChatResponse = ollama.chat(model='llama3.2', messages=[
  { 'role': 'system', 'content': 'always response in a way that\
   a 5 years old kid can understand scientific informations.', },
  { 'role': 'user', 'content': 'Why the sky is blue?', 'temperature': 0.7,},
])
print(response.message.content)
```

###openai api

```python
from openai import OpenAI

client = OpenAI(
    base_url='http://localhost:11434/v1/',

    # required but ignored
    api_key='ollama',
)

chat_completion = client.chat.completions.create(model='llama3.2', temperature=0.7, messages=[
  { 'role': 'system', 'content': 'always response in a way that\
   a 5 years old kid can understand scientific informations.', },
  { 'role': 'user', 'content': 'Why the sky is blue?', },
])


print(chat_completion.choices[0].message.content)
```

###REST API

```python
import requests

BASE_URL = "http://localhost:11434/v1/chat/completions"
HEADER = {"Content-Type": "application/json"}
DATA ={
     "model": "llama3.2", "temperature": 0.7,
      "messages": [{ 'role': 'system', 'content': 'always response in a way that\
                    a 5 years old kid can understand scientific informations.', },
                    { 'role': 'user', 'content': 'Why the sky is blue?', },
                  ]
}

response_rest = requests.post(url=BASE_URL, headers=HEADER, json=DATA)

print(response_rest.json()["choices"][0]["message"]["content"])
```

###ollama with stream=True

```python
import ollama
response_stream: ollama.ChatResponse = ollama.chat(model='llama3.2', stream=True, messages=[
  { 'role': 'system', 'content': 'always response in a way that\
   a 5 years old kid can understand scientific informations.', },
  { 'role': 'user', 'content': 'Why the sky is blue?', 'temperature': 0.7,},
])

for response in response_stream:
  print(response.message.content, end='', flush=True)
```

###ollama with async respnse

```python
import nest_asyncio
nest_asyncio.apply()
import asyncio
from ollama import AsyncClient

async def chat():
  messages = [{ 'role': 'system', 'content': 'always response in a way that\
    a 5 years old kid can understand scientific informations.', },
    { 'role': 'user', 'content': 'Why the sky is blue?', 'temperature': 0.7,}]
  response_async: ollama.ChatResponse =  await AsyncClient().chat(model='llama3.2', stream=True, messages=messages)
  async for response in response_async:
    print(response.message.content, end='', flush=True)

asyncio.run(chat())
```

#chatbot with UI (gradio)

```python
import gradio as gd
import ollama

def chat(prompt, history):

  messages = [{ 'role': 'system', 'content': 'act as professional assistant'}]
  messages.append({'role': 'user', 'content': prompt})

  response: ollama.ChatResponse =  ollama.chat(model='llama3.2', messages=messages)
  return response.message.content


gd.ChatInterface(fn=chat, type='messages', title="llama3.2 using ollama in colab",
                 examples=[{"text": "write a snake game in python"},
                           {"text": "why sky is blue?"},
                           {"text": "what is REST API?"}]).launch()
```
